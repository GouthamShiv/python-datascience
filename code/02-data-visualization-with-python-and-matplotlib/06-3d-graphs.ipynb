## `3D Graphs`

- Creating dataset
- Plotting the graph on a `3D plot`
- Customizing the graph

```python
# Importing matplotlib
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Creating data-set

import random

x = [i for i in range(1, 101)]
y = []
z = []

for i in range(100):
  y.append(random.randint(1, 101))
  z.append(random.randint(1, 101))
```

```python
# Creating 3D graph

axes = plt.axes(projection='3d')
axes.scatter3D(x, y, z)
plt.show()
```

```python
# Customizing 3D graph with labels and title

axes = plt.axes(projection='3d')
axes.scatter3D(x, y, z)
axes.set_xlabel('X Axes')
axes.set_ylabel('Y Axes')
axes.set_zlabel('Z Axes')
plt.show()
```

```python
# Customizing 3D graph with visual changes

axes = plt.axes(projection='3d')
axes.scatter3D(x, y, z, marker='^')
axes.set_xlabel('X Axes')
axes.set_ylabel('Y Axes')
axes.set_zlabel('Z Axes')
plt.show()
```
