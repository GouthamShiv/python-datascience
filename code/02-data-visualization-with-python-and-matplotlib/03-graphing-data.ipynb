### `Basic Line plot`

- Creating some random data
- Graphing the data points with a line
- Creating a basic scatter plot

```python
# Importing matplotlib
from matplotlib import pyplot as plt
```

```python
# Creating random data-set
import random

x = [i for i in range(1, 101)]
y = []
for i in range(100):
  y.append(random.randint(1, 101))
```

```python
# Plotting a basic line plot
plt.plot(x, y)
# in some cases might need to add below line for the plot to be displayed
plt.show()
```

```python
# Plotting a basic scatter plot
plt.scatter(x, y)
# in some cases might need to add below line for the plot to be displayed
plt.show()
```

### `Customizing the plots`

- Adding titles
- Changing axis
- Changing line appearance


```python
# Adding title to a plot and axises
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# Customizing the appearance of a line plot
# Colors: w r y g k c b m
plt.plot(x, y, 'g')
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# Customizing the appearance of a plot with shapes
# Colors: w r y g k c b m
# Shapes: o p s d h x v . , 1 2 3 4 8 - + | : < > ^ -- -.
plt.plot(x, y, 'rs')
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# setp() to customize, but a bit more explicit
lines = plt.plot(x, y)
plt.setp(lines, linewidth=5, color='y')
plt.show()
```
